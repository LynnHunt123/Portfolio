```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import requests
from time import time
from datetime import datetime
import os

dir_data = os.getcwd().replace("real_app", "data")
```

```python
def get_coin_price(token, days):
    print(token)
    base_url = "https://api.kucoin.com"
    coin_pair = "-".join([token, "USDT"])
    frequency = "1day" 
    now_is = int(time())
    days_delta = 60 * 60 * 24 * days
    start_At = now_is - days_delta
    price_url = f"/api/v1/market/candles?type={frequency}&symbol={coin_pair}&startAt={start_At}&endAt={now_is}"
    price_dict = {}
    prices = requests.get(base_url+price_url).json()
#     print(prices)
    for item in prices['data']:
        date_converted = datetime.fromtimestamp(int(item[0])).strftime("%Y-%m-%d")
        price_dict[date_converted] = item[1:7]
    colnames = ["open", "close", "high", "low", "volume", "amount"]
    colnames = [f"{token.lower()}_{x}" for x in colnames] 
    priceDF = pd.DataFrame(price_dict,index = colnames).T
    #Convert prices into a float
    for col in colnames:
        priceDF[col] = priceDF[col].astype(float)

    #convert dates to datetime from object
    priceDF.index = pd.to_datetime(priceDF.index)

    #reverse dates
    priceDF = priceDF.iloc[::-1]
    return priceDF

coin_tickers = ["BTC", "ETH", "XRP"] 
coins = pd.concat([get_coin_price(coin, 1500) for coin in coin_tickers], axis = 1).dropna(how = "any")
print(coins.head(5))
print(".\n.\n.\n")
print(coins.tail(5))
```

```python
# Feature engineering:

def vol_ohlc(df, ticker, lookback=10):
    if ticker is not None:
        o = df.loc[:,"_".join([ticker, "open"])]
        h = df.loc[:,"_".join([ticker, "high"])]
        l = df.loc[:,"_".join([ticker, "low"])]
        c = df.loc[:,"_".join([ticker, "close"])]
    else: 
        o = df.open
        h = df.high
        l = df.low
        c = df.close
    
    k = 0.34 / (1.34 + (lookback+1)/(lookback-1))
    cc = np.log(c/c.shift(1))
    ho = np.log(h/o)
    lo = np.log(l/o)
    co = np.log(c/o)
    oc = np.log(o/c.shift(1))
    oc_sq = oc**2
    cc_sq = cc**2
    rs = ho*(ho-co)+lo*(lo-co)
    close_vol = cc_sq.rolling(lookback).sum() * (1.0 / (lookback - 1.0))
    open_vol = oc_sq.rolling(lookback).sum() * (1.0 / (lookback - 1.0))
    window_rs = rs.rolling(lookback).sum() * (1.0 / (lookback - 1.0))
    result = (open_vol + k * close_vol + (1-k) * window_rs).apply(np.sqrt) * np.sqrt(252)
    result[:lookback-1] = np.nan
    
    return result

def gen_lags(cols, df, max_lag = 5):
    for col in cols:
        for lag in range(1, max_lag+1):
            df.loc[:,f"{col}_lag{lag}"] = df.loc[:,col].shift(-lag)
    df = df.dropna(how = "any")
    return df

# bitcoin return
df = pd.read_csv(os.path.join(dir_data, "rawdata.csv")).set_index("Unnamed: 0")
df.index.name = "Date"

df.columns = [x.lower() for x in df.columns]
print(df.columns)
for s in ["btc", "eth", "xrp"]:
    df.loc[:,f"r{s}"] = df.loc[:,f"{s}_close"].pct_change()
    
# df.loc[:,"rbtc"] = df.loc[:,"btc"].pct_change()
# spy500 return
df.loc[:,"rspy"] = df.loc[:,"adj close"].pct_change()
# change in usd-cny forex 
df.loc[:,"rusdcny"] = df.loc[:,"usdcny"].pct_change()



for series in ["btc", "eth", "xrp"]:
    # ohlc volatility
    df.loc[:,f"vol_{series}"] = vol_ohlc(df, series)
    # 14 days high
    df.loc[:,f"{series}_hh14"] = df.loc[:,f"{series}_close"].rolling(14).max()
    # 14 days low
    df.loc[:,f"{series}_ll14"] = df.loc[:,f"{series}_close"].rolling(14).min()
    # high low ratio
    df.loc[:,f"{series}_hh/ll14"] = df.loc[:,f"{series}_hh14"]/df.loc[:,f"{series}_ll14"]
    # geometric rolling range
    df.loc[:,f"{series}_grr14"] = df.loc[:,f"{series}_hh/ll14"] / df.loc[:,f"{series}_hh/ll14"].rolling(14).mean()
    
df.loc[:,"vol_spy"] = vol_ohlc(df, None)
# df.loc[:,"vol"]


df = gen_lags(cols = ["rspy", "rusdcny"] + [f"r{s}" for s in ["btc", "eth", "xrp"]], df = df, max_lag= 3)
df = df.dropna(how = "any")
df.to_csv(os.path.join(dir_data, "pricedata.csv"), 
          encoding = "utf-8", 
          index = True)
df
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import Ridge

from sklearn.model_selection import cross_validate
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, make_scorer, precision_score, accuracy_score

from sklearn.model_selection import learning_curve
import xgboost as xgb 
from xgboost import XGBRegressor, XGBRFRegressor

import pickle

def evaluate_model(model, X, y, train_size = 20, test_size = 1, 
                   scorer = make_scorer(mean_squared_error, greater_is_better=False, squared=False)
                  ):
    cv = TimeSeriesSplit(n_splits=len(X)//test_size-1, 
                         test_size = test_size, 
#                          max_train_size=train_size, 
                         gap = 0)
    
    
#     return np.mean(cross_validate(model, X, y, cv=cv, scoring=scorer, n_jobs=-1)['test_score'])
    scores = cross_validate(model, X, y, cv=cv, scoring=scorer, n_jobs=-1)['test_score']
    print("Evaluation mean: ", np.nanmean(scores))
    print("Evaluation std: ", np.nanstd(scores))
    return scores

pipeline = Pipeline([
    ('impute', SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=0.)),
    ('scale', StandardScaler()),
#     ('pca', PCA(n_components=1)),
#     ('pca', PCA()),
#     ('model', Ridge())
#     ("model", RandomForestRegressor())
    ("model", XGBRegressor(n_estimators = 100, 
                             learning_rate = 0.1, 
                             verbosity = 3, 
                             objective = "reg:squarederror", 
                             n_jobs = -1,
                            ))
])

# Best score yet using XGBoost:
# Evaluation mean:  -0.004303023022984572
# Evaluation std:  0.017450771262907148

# Best score after tuning: 
# Evaluation mean:  -0.00408149361020521
# Evaluation std:  0.016346200149106087

# Best score with grr after tuning: 
# Evaluation mean:  -0.003979630303667358
# Evaluation std:  0.015570793843213363
evaluate_model(pipeline, X, y, train_size = 50, test_size = 2)
```
